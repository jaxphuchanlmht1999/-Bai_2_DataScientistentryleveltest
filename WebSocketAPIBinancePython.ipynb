```python
import json
import time
from datetime import datetime


def on_message(ws, message):
    item = json.loads(message)    
    item['k']['time_get'] = datetime.now().strftime('%Y/%m/%d %H:%M:%S')

    print(item)
    print(" ")

    data.append(item)
    print(len(data))

    # print(message)
    time.sleep(1)
    
def on_close(ws):
    print('Connection Closed')

ws = websocket.WebSocketApp(_socket,
                            on_message = on_message,
                            on_close = on_close)
```
