```python
import json
import time
from datetime import datetime
import websocket, json

def on_message(ws, message):
    item = json.loads(message)
    item['data']['time_get'] = datetime.now().strftime('%Y/%m/%d %H:%M:%S')
    data.append(item)
    print(item)
    print(len(data))

def on_close(ws):
    print('Connection Closed')

ws = websocket.WebSocketApp(_socket,
                            on_message = on_message,
                            on_close = on_close)
```

```python
ws.run_forever()
```
