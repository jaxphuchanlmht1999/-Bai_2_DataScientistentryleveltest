```python
import queue
import threading
import time
from datetime import datetime

# The queue for tasks
# Tạo hàng chờ queue cho task
q = queue.Queue()



# Tạo Worker xử lý từng nhiệm vụ
def worker():
    while True:
        item = q.get()
        if item is None:
            break
               
        # Thời gian nghỉ giữa các Worker theo item
        time.sleep(item * PERIOD_OF_TIME)
        print("Working on " + str(item))
        print(f'Start time item {item} là ' + str(datetime.now()))

        # Thời gian bắt đầu
        start_time_get = time.time()

        global time_get_total
        time_get_total = start_time_get

        # Do Task
        # Nếu thời gian lấy > PERIOD_OF_TIME => Dừng
        time.sleep(3)
        ws.run_forever(reconnect=5)


        q.task_done()


def start_workers(worker_pool=24):
    threads = []
    for i in range(worker_pool):
        t = threading.Thread(target=worker)
        t.start()
        print('Khởi tạo threah thứ ', i)
        threads.append(t)
    return threads


def stop_workers(threads):
    # stop workers
    for i in threads:
        q.put(None)
    for t in threads:
        t.join()


def create_queue(task_items):
    for item in task_items:
        q.put(item)
```
