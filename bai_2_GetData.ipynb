```python
import websocket, json
```

```python
# cc = #'btcusdt' 'ethusdt'
cc = 'btcusdt'
interval = '1m' # 1m, 1s

socket = f'wss://stream.binance.com:9443/ws/{cc}@kline_{interval}'
```

```python
socket
```

```python
data = []
```

```python
import json
import time
from datetime import datetime


def on_message(ws, message):
    item = json.loads(message)    
    item['k']['time_get'] = datetime.now().strftime('%Y/%m/%d %H:%M:%S')

    print(item)
    print(" ")

    data.append(item)
    print(len(data))

    # print(message)
    time.sleep(1)
    
def on_close(ws):
    print('Connection Closed')

ws = websocket.WebSocketApp(socket,
                            on_message = on_message,
                            on_close = on_close)
```

```python
ws.run_forever()
```

```python
import pandas as pd


pd.DataFrame(pd.DataFrame(data)['k'])
```

```python
arr_k = []

for i in range(len(data)):
   arr_k.append(data[i]['k'])
```

```python
df_k = pd.DataFrame(arr_k)
df_k
```
